# Reflection model
それっぽいゾーンを描画してる

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (4095 - y) * np.sin(x)

x = np.linspace(0, np.pi, 500)
y = np.linspace(0, 4095, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.xticks([0, 250, 500],['0', 'π/2', 'π'])
plt.yticks([0, 500],['0', '4096'])
plt.imshow(Z, cmap='bone', origin='lower')
plt.colorbar()
plt.show()
```

```python
sigma_d = 500
d = 2000

likelihood = np.exp(-(d - Z) ** 2 / (2 * (sigma_d ** 2))) / (sigma_d * np.sqrt(2 * np.pi))

plt.xticks([0, 250, 500],['0', 'π/2', 'π'])
plt.yticks([0, 500],['0', '4096'])
plt.imshow(likelihood, cmap='bone', origin='lower')
plt.colorbar()
plt.show()
```
